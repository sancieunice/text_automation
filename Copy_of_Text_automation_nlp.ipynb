```python
!pip install gradio transformers
import gradio as gr
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

def answer_from_input(question, context):
    if not context:
        return "Please provide some context to answer questions."
    if not question:
        return "Please ask a question."
    try:
        response = qa_pipeline(question=question, context=context)
        return response['answer']
    except Exception as e:
        return f"Error generating answer: {str(e)}"

with gr.Blocks() as demo:
    gr.Markdown("# NLP Text Automation Chatbot")
    gr.Markdown("### Provide the context and ask your question.")

    context_input = gr.Textbox(
        label="Context",
        placeholder="Enter the context here (e.g., a paragraph from your notes)."
    )
    question_input = gr.Textbox(
        label="Question",
        placeholder="Ask your question based on the context."
    )
    output_answer = gr.Textbox(
        label="Answer",
        placeholder="The answer will appear here.",
        interactive=False
    )

    gr.Button("Submit").click(
        answer_from_input,
        inputs=[question_input, context_input],
        outputs=output_answer
    )

demo.launch()
```
